```python
# EDA: IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# LOAD DATASET

df = pd.read_csv("punjab_master_dataset_expanded_cleaned.csv")

# Preview data
df.head()
```

```python
# EDA STEP 2 — DISTRIBUTIONS

import matplotlib.pyplot as plt

# Yield Distribution
plt.figure()
plt.hist(df['yield_kg_per_ha'], bins=40)
plt.title("Yield Distribution")
plt.xlabel("Yield (kg/ha)")
plt.ylabel("Frequency")
plt.show()


# Rainfall Distribution
plt.figure()
plt.hist(df['rainfall'], bins=40)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency")
plt.show()


# Soil pH Distribution
plt.figure()
plt.hist(df['ph'], bins=40)
plt.title("Soil pH Distribution")
plt.xlabel("pH Value")
plt.ylabel("Frequency")
plt.show()


# Water Stress Index
plt.figure()
plt.hist(df['water_stress_index'], bins=40)
plt.title("Water Stress Index Distribution")
plt.xlabel("Water Stress Index")
plt.ylabel("Frequency")
plt.show()


# Fertility Score
plt.figure()
plt.hist(df['fertility_score'], bins=40)
plt.title("Fertility Score Distribution")
plt.xlabel("Fertility Score")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("crop")["yield_kg_per_ha"].mean().sort_values()
```

```python
# STEP 3A — CORRELATION HEATMAP


import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,10))
plt.imshow(corr)

plt.colorbar()
plt.title("Correlation Heatmap")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.tight_layout()
plt.show()
```

```python
# TOP POSITIVE CORRELATIONS

yield_corr = corr["yield_kg_per_ha"].sort_values(ascending=False)

yield_corr.head(10)
```

```python
# TOP NEGATIVE CORRELATIONS

yield_corr.tail(10)
```

```python
# STEP 4 — DISTRICT YIELD

district_yield = df.groupby("district")[
    "yield_kg_per_ha"
].mean().sort_values(ascending=False)

district_yield.head(10)
```

```python
district_yield.tail(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

district_yield.plot(kind="bar")

plt.title("Average Yield by District")
plt.ylabel("Yield (kg/ha)")
plt.xlabel("District")

plt.xticks(rotation=90)

plt.show()
```

```python
district_water = df.groupby("district")[
    "water_stress_index"
].mean().sort_values(ascending=False)

district_water.head(10)
```

```python
district_water.tail(10)
```

```python
# STEP 5 — Crop-wise Yield Analysis

# AVERAGE YIELD PER CROP

crop_yield = df.groupby("crop")[
    "yield_kg_per_ha"
].mean().sort_values(ascending=False)

crop_yield
```

```python
# CROP BY DISTRICT

crop_district = df.groupby(
    ["district", "crop"]
)["yield_kg_per_ha"].mean()

crop_district.head(20)
```

```python
# BEST CROP PER DISTRICT

best_crop = crop_district.groupby("district").idxmax()

best_crop.head(10)
```

```python
# 5D — NORMALIZED YIELD

# Normalize yield inside each crop

df["normalized_yield"] = df.groupby("crop")[
    "yield_kg_per_ha"
].transform(lambda x: (x - x.mean()) / x.std())
```

```python
# BEST CROP (CORRECTED)

crop_district_norm = df.groupby(
    ["district", "crop"]
)["normalized_yield"].mean()

best_crop_norm = crop_district_norm.groupby("district").idxmax()

best_crop_norm.head(10)
```

```python
# STEP 6A — YIELD BY SEASON

season_yield = df.groupby("season")[
    "yield_kg_per_ha"
].mean().sort_values(ascending=False)

season_yield
```

```python
# STEP 6B — CROP DISTRIBUTION

season_crop = df.groupby(
    ["season", "crop"]
).size()

season_crop
```

```python
import matplotlib.pyplot as plt

plt.figure()

season_yield.plot(kind="bar")

plt.title("Average Yield by Season")
plt.ylabel("Yield (kg/ha)")
plt.xlabel("Season")

plt.show()
```
